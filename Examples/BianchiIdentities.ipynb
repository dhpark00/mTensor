## First Kind

$$
  R_{[abc]}^{\ \ \ \ \ \ \ \ d}
  = -\nabla_{[a} t_{bc]}^{\ \ \ \ d}
  + t_{[ab}^{\ \ \ \ p} t_{c]p}^{\ \ \ \ d}
$$

```
RiemannCD[la, lb, lc, ud]
```

```
AntisymmetrizeIndices[{la, lb, lc}, %]
```

```
% // RiemannToGamma
```

```
% // GammaToMetric
```

```
lhsRc = %;
```

```
-CD[la, Torsion[lb, lc, ud]] + Torsion[la, lb, ue] Torsion[lc, le, ud]
```

## Second Kind

$$
  \nabla_{[a} R_{bc]d}^{\ \ \ \ \ \ \ e}
  = t_{[ab}^{\ \ \ \ \ p} R_{c]pd}^{\ \ \ \ \ \ \ e}
$$

```
CD[la, RiemannCD[lb, lc, ld, ue]]
```

```
AntisymmetrizeIndices[{la, lb, lc}, %]
```

```
% // CDtoGamma
```

```
% // RiemannToGamma
```

```
lhsRc = GammaToMetric[%];
```

```
Torsion[la, lb, uf] RiemannCD[lc, lf, ld, ue]
```

```
AntisymmetrizeIndices[{la, lb, lc}, %]
```

```
% // RiemannToGamma
```

```
rhsRc = GammaToMetric[%];
```

```
(* large terms and long evaluation *)
expr = lhsRc - rhsRc // TindexSort // ResetDummies // Absorbg;
```

```
Length[List @@ expr]
```

```
expr // Tsimplify
```

```
PutMetric[%, ue]
```

```
% // Tsimplify
```
