# Chapter 9. General Relativity Examples

## 9.1 Reissner-Nordstrom Invariants

### 9.1.1 Computing the Riemann Tensor and Related Objects

```
coSys = {r, \[Theta], \[Phi], t}
```

```
metric = {{1/(1 - (2 m)/r + q^2/r^2), 0, 0, 0},
           {0, r^2, 0, 0},
           {0, 0, r^2 Sin[\[Theta]]^2, 0},
           {0, 0, 0, -(1 - (2 m)/r + q^2/r^2)}}
```

### 9.1.2 Invariant Definitions

```
SetDimension[4]
```

```
invOne = ScalarCD[]^2
```

```
invTwo = SumDum[RicciCD[la, lb] RicciCD[ua, ub]]
```

```
invThree = 
 SumDum[RiemannCD[la, lb, lc, ld] RiemannCD[ua, ub, uc, ud]] // 
  TindexSort
```

```
CD[la, ua, ScalarCD[]]
```

```
invFour = SumDum[CDtoGamma[CDtoGamma[%]]];
```

```
CD[la, RicciCD[lb, lc]] CD[ua, RicciCD[ub, uc]]
```

```
invFive = SumDum[CDtoGamma[%]];
```

### 9.1.3 Evaluation of the Invariants

```
InitCTensor[coSys, metric, RicciCD -> True]
```

```
invOne
```

```
Metricg[ua, uc] Metricg[ub, ud] RicciCD[lc, ld]
```

```
ricciUp = Table[SumDum[%], {ua, 4}, {ub, 4}];
```

```
SetComponents[RicciCD[ua, ub], ricciUp]
```

```
invTwo  // Simplify
```

```
(Metricg[ua, ue] Metricg[ub, uf]
 Metricg[uc, ug] Metricg[ud, uh] RiemannCD[le, lf, lg, lh])
```

```
SetComponents[RiemannCD[ua, ub, uc, ud], riemannUp]
```

```
invThree  // Simplify
```

```
% /. q -> 0
```

```
invFour // Simplify
```

```
Metricg[ua, ud] GammaCD[ld, lb, uc]
```

```
SetComponents[GammaCD[ua, lb, uc],
               Table[SumDum[%],
                     {ua, 4}, {lb, -1, -4, -1}, {uc, 4}]]
```

```
invFive // Simplify
```

## 9.2 Variations and Field Equations

### 9.2.1 Metric Variations and the Einstein Action

mTensor 재시작:

```
<< mTensor`
```

```
Tdefine[Detg[], "g"]
```

```
lag = Sqrt[-Detg[]] Metricg[ua, ub] RicciCD[la, lb]
```

변분 $\delta \mathcal{L} / \delta g^{ab}$ 계산:

```
VD[Metricg[ua, ub], lag]
```

```
% // Absorbg
```

```
% // Tsimplify
```

부분 적분을 통해 $\delta R_{cd} / \delta g^{ab} = 0$이 된다는 사실을 지정:

```
einsteinCD = 
 1/Sqrt[-Detg[]] VD[Metricg[ua, ub], lag, IndependentVD -> {RicciCD}]
```

```
% // Absorbg
```

```
% // Tsimplify
```

부분 적분을 통한 변분 공식 $\delta g / \delta g^{ab} = -g g_{ab}$을 설정:

```
VD[Metricg[ua_, ub_], Detg[], 
  rest_, ___] := -rest Detg[] Metricg[FlipIndex[ua], FlipIndex[ub]] /;
   AllTrue[{ua, ub}, UpIndexQ]
```

```
VD[Metricg[ua, ub], Detg[]]
```

결과적으로 아인슈타인 방정식 유도:

```
einsteinCD // Absorbg
```

```
% // TindexSort
```

### 9.2.2 The Gauss-Bonnet Invariant

```
gbInv4D = 
 Sqrt[Detg[]] (RiemannCD[la, lb, lc, ld] RiemannCD[ua, ub, uc, ud] - 
    4 RicciCD[la, lb] RicciCD[ua, ub] + ScalarCD[]^2)
```

```
VD[Metricg[la, lb], %]
```

중간 생략

```
anti5g[ua_, ub_, uc_, ud_, ue_, uf_, ug_, uh_, ui_, uj_] := 
 5!*AntisymmetrizeIndices[{uf, ug, uh, ui, uj}, 
   Metricg[ua, uf] Metricg[ub, ug] Metricg[uc, uh] Metricg[ud, 
     ui] Metricg[ue, uj]]
```

```
expr = 1/8 anti5g[ua, ub, uc, ud, ue, uf, ug, uh, ui, uj] WeylCD[lb, 
   lc, li, lj] WeylCD[ld, le, lg, lh]
```

```
% // Absorbg
```

```
% // Tsimplify
```

```
weylTraceRule = 
 RuleUnique[WeylCD[a_, b_, c_, d_], 0, 
  PairIndexQ[a, b] || PairIndexQ[a, c] || PairIndexQ[a, d] || 
   PairIndexQ[b, c] || PairIndexQ[b, d] || PairIndexQ[c, d]]
```

```
expr = %% /. weylTraceRule
```
