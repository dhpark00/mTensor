## 6. Fix Bug in `xTensor`

```
S[lb, uc] F[ub, lc]
```

```
ToCanonical[%]
```

```
S[lb, lc] LD[v, F[ub, uc]]
```

```
ToCanonical[%]
```

```
S[lb, uc] LD[v, F[ub, lc]]
```

As an operator $\cal{L}_v$ for indexed tensors,
$$
  \cal{L}_v T_a^{\ b} \textrm{ == } \cal{L}_v (T_a^{\ b})
$$ $$
  \cal{L}_v T_a^{\ b}  \textrm{ =!= } (\cal{L}_v T)_a^{\ b}
$$

```
ToCanonical[%]  (* fixed a Bug in xTensor *)
```

```
ToCanonical[%%, CovDs -> {CD, LD}]  (* OK! *)
```
