### ToCycl (PermutationCycles) and ToImag (PermutationList)

```
PermutationCycles[{4, 3, 1, 2, 7, 6, 5}]
```

`Imag` 표기는 길이에 대한 정보를 포함하지만 `Cycles` 표기는 길이 정보를 포함하지 않으므로, 함수에 `Cycles`를 전달할 때는 길이를 함수의 인자로 전달해야 한다. 길이 인자가 없으면 `permDeg`를 이용하여 구한 값을 길이로 사용한다. 

```
PermutationList[Cycles[{{1, 2, 3}, {4, 7}}]]
```

```
ToImag[{Cycles[{{1, 2, 3}, {4, 7}}], E^(I Re[\[Theta]])}, 8]
```

```
ToImag[{{1, 2, 3}, {4, 7}}, 8]  (* from cycl_List *)
```

```
ToImag[{Cycles[{{1, 2, 3}, {4, 7}}], E^(I Re[\[Theta]])}]
```

```
ToImag[{Cycles[{{1, 4, 2, 3}, {5, 7}}], E^(I Re[\[Theta]])}]
```
