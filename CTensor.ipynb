## 1. InitCTensor

### 1.1 Initialization of Component Tensors in Coordinate Basis

좌표계와 계량 텐서를 입력으로 받아 Coordinate Basis에서 성분 텐서를 초기화 한다. 옵션으로 `SimplifyMore`, `Verbose`, `GammaCD`, `RiemannCD`, `RicciCD`가 있다.

```
coSys = {t, x};
```

```
metric = {{-E^(2 \[Rho][t, x]), 0}, {0, E^(
   2 \[Rho][t, x])}}; MatrixForm[metric]
```

```
InitCTensor[coSys, metric, RicciCD -> True]
```

```
Status[]
```

```
Show[Metricg]
```

```
Show[Structuref]
```

```
Show[GammaCD]
```

```
Show[RiemannCD]
```

```
Show[RicciCD]
```

```
Show[ScalarCD]
```

### 1.2 Initialization of Component Tensor in Non-Coordinate Basis

좌표계, metric, 기준벡터의 행렬표현을 입력으로 받아 Non-coordinate Basis에서 성분 텐서를 초기화 한다.

```
coSys = {t, x};
```

Orthonomal Basis $(e_a = h_a^\mu \partial_\mu)$

```
metric = {{-1, 0}, {0, 1}};
```

```
basisM = {{E^-\[Rho][t, x], 0}, {0, 
   E^-\[Rho][t, x]}}; MatrixForm[basisM]
```

```
InitCTensor[coSys, metric, basisM, RicciCD -> True]
```

## 2. Calculations of Component Tensors

### 2.1 Coordinates, Metric, and InitCTensor

`Tcalc` 함수의 인자로 `Structuref`, `GammaCD`, `RiemannCD`, `RicciCD`, `ScalarCD`, `WeylCD` 등을 사용하여 $f_{ab}^{\ \ \ c}$, $\Gamma_{ab}^{\ \ \ c}$, $R_{abc}^{\ \ \ \ \ d}$, $R_{ab}$, $R$, $C_{abc}^{\ \ \ \ \ d}$를 각각 계산한다. 옵션으로는 simplification을 위한 함수의 이름이 될 수 있다. 이 함수를 사용하려면 `InitCTensor` 함수를 이용하여 성분 텐서를 먼저 초기화시켜야 한다.

Coordinate Systems, Options, and Metric for the Schwarzschild SpaceTime

```
coSys = {t, r, \[Theta], \[Phi]};
```

```
SetOptions[Together, Trig -> True]; (* for default Csimplify *)
```

```
metric = {{-(1 - (2 G M)/r), 0, 0, 0}, {0, 1/(1 - (2 G M)/r), 0, 
    0}, {0, 0, r^2, 0}, {0, 0, 0, r^2 Sin[\[Theta]]^2}};
```

```
InitCTensor[coSys, metric, Verbose -> True]
```

## 3. Commands in CTensor

### 3.1 Geodesic

시공간의 geodesic을 얻는다. 인자는 정수로 좌표계의 순서이다. 옵션으로 simplification을 위한 함수가 될 수 있다. 이 함수를 사용하려면 `InitCTensor` 함수를 이용하여 성분 텐서를 초기화하여야 한다. 

```
coSys = {t, r, \[Theta], \[Phi]};
```

```
metric = {{-(1 - (2 G M)/r), 0, 0, 0}, {0, 1/(1 - (2 G M)/r), 0, 
    0}, {0, 0, r^2, 0}, {0, 0, 0, r^2 Sin[\[Theta]]^2}};
```

```
InitCTensor[coSys, metric]
```

```
Geodesic[1, Simplify]
```

```
Geodesic[2, Simplify]
```

```
Geodesic[3, Simplify]
```

```
Geodesic[4, Simplify]
```
